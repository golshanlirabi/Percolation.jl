# How to use

To use this percolation package, you should do Pkg.clone("https://github.com/tkMeitner/Percolation.jl").
This package use PyPlot to plot figure and ImageMagick to make gif animation.

```julia
using Percolation
```

# Nearest Neighbor

```julia
# N: lattice size
# p: occupied probability
N = 100; M = N; p = 0.6

# make N by M lattice and occupied each site with probability p. But in version 0 only support N by N lattice.
# I will support for N by M lattice in future.
square = squarenn(N, M, p)
percolation(square, colorbar=false, color="hot")
```

# next nearest neighbor

```julia
# If you want to simulate not only nearest neighbor but next nearest neighbor, use squarennn type.
N = 100; p = 0.45
size = squarennn(N, N, p)
percolation(size)
```

## Order Parameter

```julia
using PyPlot
```

```julia
plist = collect(0:0.01:1)
hit = zeros(Int64, length(plist))
N = 50; trial = 10
for i in 1:length(plist)
    for itr in 1:trial
        simple = squarenn(N, N, plist[i])
        hit[i] += percolation(simple, fig=false, colorbar=false, water=false)[1];
    end
end
hit /= trial;
plot(plist, hit);
axis([0, 1, -0.1, 1.1]);
```
