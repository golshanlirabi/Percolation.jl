## N dimensional lattice

```julia
N = 20; dim = 3; p = 0.4;
site = simplenn(N, dim, p);
percolation(site)
```

```julia
using PyPlot
```

```julia
plist = collect(0:0.05:1)
hit = zeros(Int, length(plist))
trial = 10; N = 10; dim = 3
for i in 1:length(plist)
    for j in 1:trial
        site = simplenn(N, dim, plist[i])
        hit[i] += percolation(site, fig=false, water=false, colorbar=false)[1]
    end
end
plot(plist, hit/trial)
```
